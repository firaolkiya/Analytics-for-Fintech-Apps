```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read
```

```python
# upload prepered data
df = pd.read_csv("/content/reviews_with_sentiment.csv")
```

**Sentiment Distribution by Bank**

```python
print("Sentiment Distribution by Bank\n\n")
sns.countplot(data=df, x='sentiment_label', hue='bank')
plt.title('Sentiment Distribution by Bank')
plt.xlabel('Sentiment')
plt.ylabel('Review Count')
plt.xticks(rotation=0)
plt.tight_layout()
```

**⭐ Rating Distribution (Histogram)**

```python
print("⭐ Rating Distribution (Histogram)\n\n")
sns.histplot(data=df, x='rating', hue='bank', multiple='stack', bins=5)
plt.title('Rating Distribution Across Banks')
plt.xlabel('Rating')
plt.ylabel('Review Count')
plt.tight_layout()
```

**☁️ WordCloud per Bank (Theme Visualization)**

```python
print("☁️ WordCloud per Bank (Theme Visualization)\n\n")
from wordcloud import WordCloud
import matplotlib.pyplot as plt

for bank in df['bank'].unique():
    text = ' '.join(df[df['bank'] == bank]['review'].dropna().astype(str).values)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {bank}')
    plt.show()
```

**Sentiment Score Trend**

```python
print(" Sentiment Score Trend\n\n")

df['date'] = pd.to_datetime(df['date'])
sns.lineplot(data=df, x='date', y='sentiment_score', hue='bank')
plt.title('Sentiment Score Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
```
